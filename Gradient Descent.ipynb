```python
# Importing libraries which will be in use
import numpy as np
import pandas as pd
```

```python
# Creating an array for slope(m) and intercept(c)
a = np.array([92,56,88,80,49,65,35,66,67])
b = np.array([98,68,81,83,52,66,30,68,73])
data = pd.DataFrame()
```

```python
# Creating a Cost function with learning rate of 0.005 
def gradient_descent(x,y):
    
    m_curr = c_curr = 0
    iterations = 60
    n = len(x)
    learning_rate = 0.01
    cost_all = []
    m_all=[]
    c_all = []
    
    for i in range(iterations):
        
        y_pred = m_curr * x + c_curr
        m_der = -(2/n)*sum(x*(y-y_pred))
        c_der = -(2/n)*sum((y-y_pred))
        cost = (1/n) * sum([val**2 for val in (y-y_pred)])
        m_curr = m_curr + learning_rate * m_der
        c_curr = c_curr + learning_rate * c_der
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,c_curr,cost, i))
        cost_all.append(cost)
        m_all.append(m_curr)
        c_all.append(c_curr)

    data['costs'] = cost_all
    data['slope'] = m_all
    data['constant'] = c_all



    
```

```python
gradient_descent(a,b)
```

```python
data.sort_values(by='costs', ascending=True)
```
